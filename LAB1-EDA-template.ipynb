## Introducción

El **Exploratory Data Analysis (EDA)** consiste en analizar un conjunto de datos para comprender sus características principales mediante **estadística descriptiva y visualización**.

El objetivo del EDA es:

- Entender la estructura de los datos  
- Identificar patrones temporales y estacionales  
- Detectar anomalías o problemas de calidad  
- Extraer conocimiento antes de aplicar modelos

> El EDA debe contar una historia basada en los datos.
